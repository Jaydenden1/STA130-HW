```python
# Q2
import numpy as np

# Set up initial variables
doors = [1, 2, 3]  # Three doors
wins = 0  # Count the number of wins
reps = 100000  # Number of simulations

for _ in range(reps):
    winning_door = np.random.choice(doors)  # Randomly choose the winning door
    my_choice = np.random.choice(doors)  # Player's initial choice

    # Simulate Monty revealing a goat door (Monty won't reveal the player's choice or the winning door)
    available_doors = [door for door in doors if door != my_choice and door != winning_door]
    monty_reveal = np.random.choice(available_doors)  # Monty reveals a goat door

    # Player switches: choose the remaining door that hasn't been revealed
    switch_choice = [door for door in doors if door != my_choice and door != monty_reveal][0]

    # If the switched door is the winning door, count it as a win
    if switch_choice == winning_door:
        wins += 1

# Calculate the winning probability with the switch strategy
print(f"Winning probability with the switch strategy: {wins/reps}")
# Simplified Explanation of the Code:
# Initial Variables: Three doors [1, 2, 3], wins to record the number of wins, and reps to define the number of simulations.
# Each Simulation: Inside the for loop, a winning door and the player’s initial choice are randomly selected.
# Reveal a Goat Door: Using list comprehension, we exclude the winning door and the player’s choice to let Monty reveal a door with a goat.
# Switch Door: The player chooses the remaining door that hasn’t been revealed.
# Calculate Win Rate: If the switched door is the winning door, wins increments by 1. Finally, the code outputs the winning probability of the switch strategy.

#The optimized Monty Hall code is simpler and easier to understand because it eliminates unnecessary complexity. By using list comprehensions, it clearly shows the logic of removing doors without confusing try-except blocks. The code follows a straightforward sequence, making each step easy to follow and explain. It reduces the number of operations and presents the problem in a clear, readable way. This makes it more approachable for students or anyone trying to grasp the concept quickly and efficiently.
```
